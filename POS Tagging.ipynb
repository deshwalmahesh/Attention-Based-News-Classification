###  We'll be using the very simple approach to find out POS using NLTK
Building an Encode-Decoder model specially for the dataset would not have been feasible in the time frame.

```python
import nltk 
# from nltk.corpus import stopwords  # We'll include stp words too
from nltk.tokenize import word_tokenize, sent_tokenize, PunktSentenceTokenizer
# Try out different tokenizers

import pandas as pd

from sklearn.datasets import fetch_20newsgroups
```

```python
article = fetch_20newsgroups()['data'][0]
print(article)
```

```python
only_article = article.split('\n\n')[1] # split on multiple lines to get only the article
print(only_article)
```

```python
tags = []
for sent in sent_tokenize(only_article):
    tags.extend(nltk.pos_tag(word_tokenize(sent)))
```

```python
df = pd.DataFrame(tags,columns=['WORD','TAG'])
df.head()
```
